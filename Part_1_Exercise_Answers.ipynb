```python
# initialize dataframes to start where we left of with the instructions
#
import numpy as np
import pandas as pd
df = pd.read_excel('DirPrint_Filelist - clean.xlsx')
df
```

```python
df = df.set_index("LastMod")
```

## 1 Use a regular expression to select office files based on file extension

```python
# here is a simple example using a regular expression to find all rows with xls in the Ext colum
df[df.Ext.str.match(r'(xls|doc|xlsx|docx|ppt|pptx)')==True]
```

## 2 Remove duplicate entries in df based on the md5 value. 

```python
df.shape
```

```python
df_dedup = df.drop_duplicates('Md5', keep='last')
df_dedup.shape
```

## 3 For a given file, find if it has duplicates and list the folders where duplicates are located

```python
df.duplicated('Md5')
```

```python
# Apparently record 255826 is duplicated. We can print it's Md5 value using the df.loc attribute
#
df.loc[255827].Md5
```

```python
# We can select all duplicates with a filter
#
df[df.Md5 == '8280BBD8640ECF3826460903760D9774']
```

```python
df[df.Size==0]
```

## 4 Create a pivot table showing number of files per month based on data last modified

```python
df.pivot_table(index=[df.index.year,df.index.month],values='Size',aggfunc=['count','sum'])
```

## 5 Create a pivot table listing number of files vs file extension

```python
df.pivot_table(index='Ext',aggfunc=['count'])
```

```python
# Alternative count like this:

df.groupby('Ext').count()
```
